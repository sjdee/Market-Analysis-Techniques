```
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import classification_report
import numpy as np
import warnings
warnings.filterwarnings("ignore")


def run_classifier (pDf,minify_data,less_columns,print_data):  

  
  df = pDf
  
  if(less_columns==True):
    df = df.loc[:,['BEST_ANALYST_RATING', 'RETURN_ON_ASSET', 'BEST_TARGET_PRICE', 'CUR_MKT_CAP', 'SHORT_INT', 'TOT_BUY_REC', 'TOT_SELL_REC', 'day1','day2', 'day3', 'day4','day5', 'day6','day7', 'day8', 'day9', 'day10']]
    df.dropna(inplace=True)
    features = df.iloc[:,:-10]
    labels = df.iloc[:,-10:]
    
  else:
    df.dropna(inplace=True)
    features = df.iloc[:,5:-12]
    labels = df.iloc[:,-10:]
       
    
  report_data = []


  for i in range(len(labels.columns)):
    # specify the feature set, target set, the test size and random_state to select records randomly
    X_train, X_test, y_train, y_test = train_test_split(features, labels.iloc[:,i], test_size=0.3,random_state=0) 

    # Scaling values in the feature set
    scaling = MinMaxScaler(feature_range=(0,1)).fit(X_train)
    X_train = scaling.transform(X_train)
    X_test = scaling.transform(X_test)

    #Create KNN Classifier with different parameters
    clf5 = KNeighborsClassifier(n_neighbors=5)
    clf7 = KNeighborsClassifier(n_neighbors=7)
    clf9 = KNeighborsClassifier(n_neighbors=12)

    # Train the model using the training sets
    clf5.fit(X_train, y_train)
    clf7.fit(X_train, y_train)
    clf9.fit(X_train, y_train)

    # Predict the response for test dataset
    y_pred5 = clf5.predict(X_test)
    y_pred7 = clf7.predict(X_test)
    y_pred9 = clf9.predict(X_test)

    
     
    accuracy_5 = metrics.accuracy_score(y_test, y_pred5)
    report_5 = classification_report(y_test, y_pred5)
    f1_scores_5 =[] 
    f1_scores_5.insert(0, metrics.f1_score(y_test, y_pred5, average='macro'))
    f1_scores_5.insert(1, metrics.f1_score(y_test, y_pred5, average='micro'))
    f1_scores_5.insert(2, metrics.f1_score(y_test, y_pred5, average='weighted'))
    accumulate_data(report_data,report_5,accuracy_5,labels.iloc[:,i].name,"KNN (5 neighbours)",minify_data,f1_scores_5)

    
    
    
    accuracy_7 = metrics.accuracy_score(y_test, y_pred7)
    report_7 = classification_report(y_test, y_pred7)
    f1_scores_7 =[] 
    f1_scores_7.insert(0, metrics.f1_score(y_test, y_pred7, average='macro'))
    f1_scores_7.insert(1, metrics.f1_score(y_test, y_pred7, average='micro'))
    f1_scores_7.insert(2, metrics.f1_score(y_test, y_pred7, average='weighted'))
    accumulate_data(report_data,report_7,accuracy_7,labels.iloc[:,i].name,"KNN (7 neighbours)",minify_data,f1_scores_7)

    
    
    accuracy_9 = metrics.accuracy_score(y_test, y_pred9)
    report_9 = classification_report(y_test, y_pred9)
    f1_scores_9 =[] 
    f1_scores_9.insert(0, metrics.f1_score(y_test, y_pred9, average='macro'))
    f1_scores_9.insert(1, metrics.f1_score(y_test, y_pred9, average='micro'))
    f1_scores_9.insert(2, metrics.f1_score(y_test, y_pred9, average='weighted'))
    accumulate_data(report_data,report_9,accuracy_9,labels.iloc[:,i].name,"KNN (9 neighbours)",minify_data,f1_scores_9)

    

  return report_data       
```

```
def accumulate_data(report_data,report,accuracy,day_name,model_name,minify_data,f1_scores):
      
  print(day_name)
  print(report)
  
  if(minify_data == True):

    row = {}
    row['day'] = day_name.replace("day", "") 
    row['accuracy'] = accuracy
    row['f1score_macro'] = f1_scores[0]
    row['f1score_micro'] = f1_scores[1]
    row['f1score_weigthed'] = f1_scores[2]
    row['model'] = model_name

    # unravel report for the given day       
    lines = report.split('\n')
    for line in lines[2:-5]:

      row_data = line.split('     ')

      # update recall for sell
      if(float(row_data[1])==0.0):
        row['sell_recall']= float(row_data[3])
      # update precison for buy
      if(float(row_data[1])==2.0):
        row['buy_precison']= float(row_data[2])

    report_data.append(row)


  else:    
    lines = report.split('\n')
    for line in lines[2:-5]:
        row = {}
        row_data = line.split('     ')
        row['model'] = model_name
        row['accuracy'] = accuracy
        row['day'] = labels.iloc[:,i].name
        row['class'] = row_data[1]
        row['precision'] = float(row_data[2])
        row['recall'] = float(row_data[3])
        row['f1_score'] = float(row_data[4])
        row['support'] = float(row_data[5])
        report_data.append(row)

  return report_data
```
